```python
duplicate = data[data.duplicated(['cti_ext'], keep=False)]
```

```python
def create_pharmprod(data):

    artifact={}
    artifact["resourceType"]="MedicationKnowledge"
    artifact["meta"]= {
    "profile": [
      "http://hl7.org/fhir/us/example/StructureDefinition/PharmaceuticalProduct"
    ]
  }

    artifact["status"]="active"
    artifact["domain"]="human" #hardcoded for now

    artifact["synonym"]=[data["vmpgroupname"] ]
    
  
    artifact["ingredient"]=[]
    denominator={"value":1} #must have 1 for ratio...
    if idx in duplicate.index: ##assocation
      #Association ids
      id_ass=duplicate.loc[idx]["cti_ext"]
      for idx2,data2 in duplicate[duplicate["cti_ext"]==id_ass].iterrows():
          print(data2)
          ingredient={}

          ingredient["itemCodeableConcept"]={ "system":"http://www.belgium.be/ingredient","value":0,"display":data2["basis"].strip()}
          ingredient["isActive"]=True
         
          split_num=data["strength"].split(" ")
          numerator={"value":float(split_num[0]),"unit":data["strength2"].strip()}
          strength={"numerator":numerator}
          if data["denomstrength"]!=" ":
            print("...."*40)
            print(data["denomstrength"])
            split_denom=data["denomstrength"].split(" ")

            denominator={"value":float(split_denom[0]),"unit":split_denom[1].strip()}
          strength["denominator"]=denominator
          
          ingredient["strength"]=strength

          if not any([ingredient==ing for ing in artifact["ingredient"]]): #check if ingredient already exists
            artifact["ingredient"].append(ingredient)
    else: #standalone
      ingredient={}

      ingredient["isActive"]=True
      split_num=data["strength"].split(" ")
      numerator={"value":float(split_num[0]),"unit":data["strength2"].strip()}
      strength={"numerator":numerator}
        
      if data["denomstrength"]=="":
        print(data["denomstrength"])
        denominator={"value":int(data["denominator_value"]),"unit":data["denominator_unit"].strip()}
      strength["denominator"]=denominator
      ingredient["strength"]=strength
      artifact["ingredient"].append(ingredient)

    artifact["doseForm"] ={ ##create link with snomed or idmp
    "coding": [
      {
        "code": data["edqmformid"],
        "system": "http://standardterms.edqm.eu",
        "display": data["edqmform"]
      }
    ]
  }
    route=[]
   
    route.append( {
      "coding": [
        {
          "code": data["edqmid"],
          "system": "http://standardterms.edqm.eu",
          "display": data["edqmdoseform"]
        }
      ]
    })
    artifact["intendedRoute"]=route

    #time.sleep(0.1)
    #print(artifact)
    artifact_list.append(artifact) 
  return artifact_list
```
